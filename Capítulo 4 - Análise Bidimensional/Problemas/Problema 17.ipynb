### Faça o gráfico q x q para as variáveis salário de professor secundário e salário de administrador do CD-Salários.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('C:\Statistics-101\Conjunto de Dados.xls', sheet_name='CD-Salarios', header=0, skiprows=[0,1,2,3])
df
```

```python
df_describe=df.describe(percentiles=[0.25, 0.5, 0.75])
df_describe
```

```python
plt.scatter(df['Prof. Sec.'].sort_values(axis=0, kind='quicksort'), df['Administrador'].sort_values(axis=0, kind='quicksort'), c='Orange')

plt.xlim(0, df_describe.at['max', 'Prof. Sec.']*1.05)
plt.ylim(0, df_describe.at['max', 'Administrador']*1.05)

plt.show()
```
