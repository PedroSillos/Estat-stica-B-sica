## Problema 9

#### A companhia A de dedetização afirma que o processo por ela utilizado garante um efeito mais prolongado do que aquele obtido por seus concorrentes mais diretos. Veja a tabela. Você acha que existe alguma evidência a favor ou contra a afirmação feita pela Companhia A?

```python
import pandas as pd
import statistics as stat
import math
```

```python
df_frequencia_conjunta=pd.DataFrame(
    data={
        "Menos_4_meses":[64, 104, 27, 64+104+27],
        "4_a_8_meses":[120, 175, 48, 120+175+48],
        "Mais_8_meses":[16, 21, 5, 16+21+5],
        "Total":[64+120+16, 104+175+21, 27+48+5, 195+343+42]
    },
    index=['Cia_A','Cia_B','Cia_C','Total']
)
df_frequencia_conjunta
```

```python
df_frequencia_conjunta_proporcao_observada=pd.DataFrame(
    data={
        "Menos_4_meses":[
            [
                df_frequencia_conjunta.at['Cia_A','Menos_4_meses'],
                df_frequencia_conjunta.at['Cia_A','Menos_4_meses']/df_frequencia_conjunta.at['Cia_A','Total']
            ],
            [
                df_frequencia_conjunta.at['Cia_B','Menos_4_meses'],
                df_frequencia_conjunta.at['Cia_B','Menos_4_meses']/df_frequencia_conjunta.at['Cia_B','Total']
            ],
            [
                df_frequencia_conjunta.at['Cia_C','Menos_4_meses'],
                df_frequencia_conjunta.at['Cia_C','Menos_4_meses']/df_frequencia_conjunta.at['Cia_C','Total']
            ],
            [
                df_frequencia_conjunta.at['Total','Menos_4_meses'],
                df_frequencia_conjunta.at['Total','Menos_4_meses']/df_frequencia_conjunta.at['Total','Total']
            ]
        ],
        "4_a_8_meses":[
            [
                df_frequencia_conjunta.at['Cia_A','4_a_8_meses'],
                df_frequencia_conjunta.at['Cia_A','4_a_8_meses']/df_frequencia_conjunta.at['Cia_A','Total']
            ],
            [
                df_frequencia_conjunta.at['Cia_B','4_a_8_meses'],
                df_frequencia_conjunta.at['Cia_B','4_a_8_meses']/df_frequencia_conjunta.at['Cia_B','Total']
            ],
            [
                df_frequencia_conjunta.at['Cia_C','4_a_8_meses'],
                df_frequencia_conjunta.at['Cia_C','4_a_8_meses']/df_frequencia_conjunta.at['Cia_C','Total']
            ],
            [
                df_frequencia_conjunta.at['Total','4_a_8_meses'],
                df_frequencia_conjunta.at['Total','4_a_8_meses']/df_frequencia_conjunta.at['Total','Total']
            ]
        ],
        "Mais_8_meses":[
            [
                df_frequencia_conjunta.at['Cia_A','Mais_8_meses'],
                df_frequencia_conjunta.at['Cia_A','Mais_8_meses']/df_frequencia_conjunta.at['Cia_A','Total']
            ],
            [
                df_frequencia_conjunta.at['Cia_B','Mais_8_meses'],
                df_frequencia_conjunta.at['Cia_B','Mais_8_meses']/df_frequencia_conjunta.at['Cia_B','Total']
            ],
            [
                df_frequencia_conjunta.at['Cia_C','Mais_8_meses'],
                df_frequencia_conjunta.at['Cia_C','Mais_8_meses']/df_frequencia_conjunta.at['Cia_C','Total']
            ],
            [
                df_frequencia_conjunta.at['Total','Mais_8_meses'],
                df_frequencia_conjunta.at['Total','Mais_8_meses']/df_frequencia_conjunta.at['Total','Total']
            ]
        ],
        "Total":[
            [
                df_frequencia_conjunta.at['Cia_A','Total'],
                df_frequencia_conjunta.at['Cia_A','Total']/df_frequencia_conjunta.at['Cia_A','Total']
            ],
            [
                df_frequencia_conjunta.at['Cia_B','Total'],
                df_frequencia_conjunta.at['Cia_B','Total']/df_frequencia_conjunta.at['Cia_B','Total']
            ],
            [
                df_frequencia_conjunta.at['Cia_C','Total'],
                df_frequencia_conjunta.at['Cia_C','Total']/df_frequencia_conjunta.at['Cia_C','Total']
            ],
            [
                df_frequencia_conjunta.at['Total','Total'],
                df_frequencia_conjunta.at['Total','Total']/df_frequencia_conjunta.at['Total','Total']
            ]
        ]
    },
    index=['Cia_A','Cia_B','Cia_C','Total']
)
print('Distribuição conjunta de frequência e proporções observadas:')
df_frequencia_conjunta_proporcao_observada
```

Essas proporções indicam que não há diferenças consideráveis da duração de efeito de dedetização entre as três
empresas.
