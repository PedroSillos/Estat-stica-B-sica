### Calcule o grau de associação entre as variáveis estado civil e idade, na Tabela 2.1.

```python
import pandas as pd
import statistics as stat
import math as math
```

```python
df = pd.read_excel (r'C:\Statistics-101\Conjunto de Dados.xls', sheet_name='Tabela 2.1')
df
```

```python
print(df["Idade - Anos"].describe())

soma_categoria_solteiro=0
soma_categoria_casado=0
count_categoria_solteiro=0
count_categoria_casado=0

for index, row in df.iterrows():
    if row['Estado Civil'] == 'solteiro':
        soma_categoria_solteiro+=row['Idade - Anos']
        count_categoria_solteiro+=1        
    if row['Estado Civil'] == 'casado':
        soma_categoria_casado+=row['Idade - Anos']
        count_categoria_casado+=1

media_categoria_solteiro=soma_categoria_solteiro/count_categoria_solteiro
media_categoria_casado=soma_categoria_casado/count_categoria_casado  

print('\nCount categoria solteiro:', count_categoria_solteiro)
print('Count categoria casado:', count_categoria_casado)
print('\nSoma categoria solteiro:', soma_categoria_solteiro)
print('Soma categoria casado:', soma_categoria_casado)
print('\nMédia categoria solteiro:', media_categoria_solteiro)
print('Média categoria casado:', media_categoria_casado)

variancia_categoria_solteiro=0
variancia_categoria_casado=0

for index, row in df.iterrows():
    if row['Estado Civil'] == 'solteiro':
        variancia_categoria_solteiro+=( ( row['Idade - Anos']-media_categoria_solteiro )**2 )/count_categoria_solteiro
    if row['Estado Civil'] == 'casado':
        variancia_categoria_casado+=( ( row['Idade - Anos']-media_categoria_casado )**2 )/count_categoria_casado
        
print('\nVariância categoria solteiro:', variancia_categoria_solteiro)
print('Variância categoria casado:', variancia_categoria_casado)

media_das_variancias=( (count_categoria_solteiro*variancia_categoria_solteiro)+(count_categoria_casado*variancia_categoria_casado) )/(count_categoria_solteiro+count_categoria_casado)

print('\nMédia das variâncias:', media_das_variancias)

variancia_global=stat.pvariance(df['Idade - Anos'])

print('\nVariancia global:',variancia_global)

Grau_associacao_estado_civil_idade = 1 - (media_das_variancias/variancia_global)
print('\nGrau de associação entre as variáveis estado civil e idade:', Grau_associacao_estado_civil_idade)
```
