#### Usando os dados do Problema 1, calcule o valor do X² e o coeficiente de contingência C. Esses valores estão de acordo com as conclusões obtidas anteriormente?

```python
import pandas as pd
import statistics as stat
import math
```

```python
df=pd.read_excel('C:\Estatística Básica\Conjunto de Dados.xls', sheet_name='Tabela 2.1', header=0)
```

```python
frequencias_conjuntas = {}
frequencias_conjuntas['Fundamental*Capital']=0
frequencias_conjuntas['Fundamental*Interior']=0
frequencias_conjuntas['Fundamental*Outra']=0
frequencias_conjuntas['Médio*Capital']=0
frequencias_conjuntas['Médio*Interior']=0
frequencias_conjuntas['Médio*Outra']=0
frequencias_conjuntas['Superior*Capital']=0
frequencias_conjuntas['Superior*Interior']=0
frequencias_conjuntas['Superior*Outra']=0

for index, row in df.iterrows():
    if row['Grau de Instrução']=='ensino fundamental':
        if row['Região de Procedência']=='capital':
            frequencias_conjuntas['Fundamental*Capital']+=1
        if row['Região de Procedência']=='interior':
            frequencias_conjuntas['Fundamental*Interior']+=1
        if row['Região de Procedência']=='outra':
            frequencias_conjuntas['Fundamental*Outra']+=1
    if row['Grau de Instrução']=='ensino médio':
        if row['Região de Procedência']=='capital':
            frequencias_conjuntas['Médio*Capital']+=1
        if row['Região de Procedência']=='interior':
            frequencias_conjuntas['Médio*Interior']+=1
        if row['Região de Procedência']=='outra':
            frequencias_conjuntas['Médio*Outra']+=1
    if row['Grau de Instrução']=='superior':
        if row['Região de Procedência']=='capital':
            frequencias_conjuntas['Superior*Capital']+=1
        if row['Região de Procedência']=='interior':
            frequencias_conjuntas['Superior*Interior']+=1
        if row['Região de Procedência']=='outra':
            frequencias_conjuntas['Superior*Outra']+=1

df_frequencia_conjunta=pd.DataFrame(
    data={
        "Fundamental":[frequencias_conjuntas['Fundamental*Capital'],
                       frequencias_conjuntas['Fundamental*Interior'],
                       frequencias_conjuntas['Fundamental*Outra'],
                       frequencias_conjuntas['Fundamental*Capital']+frequencias_conjuntas['Fundamental*Interior']+frequencias_conjuntas['Fundamental*Outra']
                      ],
        "Médio":[frequencias_conjuntas['Médio*Capital'],
                 frequencias_conjuntas['Médio*Interior'],
                 frequencias_conjuntas['Médio*Outra'],
                 frequencias_conjuntas['Médio*Capital']+frequencias_conjuntas['Médio*Interior']+frequencias_conjuntas['Médio*Outra']
                ],
        "Superior":[frequencias_conjuntas['Superior*Capital'],
                    frequencias_conjuntas['Superior*Interior'],
                    frequencias_conjuntas['Superior*Outra'],
                    frequencias_conjuntas['Superior*Capital']+frequencias_conjuntas['Superior*Interior']+frequencias_conjuntas['Superior*Outra']
                ],
        "Total":[frequencias_conjuntas['Fundamental*Capital']+frequencias_conjuntas['Médio*Capital']+frequencias_conjuntas['Superior*Capital'],
                 frequencias_conjuntas['Fundamental*Interior']+frequencias_conjuntas['Médio*Interior']+frequencias_conjuntas['Superior*Interior'],
                 frequencias_conjuntas['Fundamental*Outra']+frequencias_conjuntas['Médio*Outra']+frequencias_conjuntas['Superior*Outra'],
                 len(df)
                ]
    },
    index=['Capital','Interior','Outra','Total']
)
print('Distribuição de frequência conjunta:')
df_frequencia_conjunta
```

```python
df_distribuicao_proporcoes_grau_educacao_segundo_cada_uma_regioes_de_procedência=pd.DataFrame(
    data={
        "Fundamental":[
            df_frequencia_conjunta.at['Capital','Fundamental']/df_frequencia_conjunta.at['Capital','Total'],
            df_frequencia_conjunta.at['Interior','Fundamental']/df_frequencia_conjunta.at['Interior','Total'],
            df_frequencia_conjunta.at['Outra','Fundamental']/df_frequencia_conjunta.at['Outra','Total'],
            df_frequencia_conjunta.at['Total','Fundamental']/df_frequencia_conjunta.at['Total','Total']
        ],
        "Médio":[
            df_frequencia_conjunta.at['Capital','Médio']/df_frequencia_conjunta.at['Capital','Total'],
            df_frequencia_conjunta.at['Interior','Médio']/df_frequencia_conjunta.at['Interior','Total'],
            df_frequencia_conjunta.at['Outra','Médio']/df_frequencia_conjunta.at['Outra','Total'],
            df_frequencia_conjunta.at['Total','Médio']/df_frequencia_conjunta.at['Total','Total']
        ],
        "Superior":[
            df_frequencia_conjunta.at['Capital','Superior']/df_frequencia_conjunta.at['Capital','Total'],
            df_frequencia_conjunta.at['Interior','Superior']/df_frequencia_conjunta.at['Interior','Total'],
            df_frequencia_conjunta.at['Outra','Superior']/df_frequencia_conjunta.at['Outra','Total'],
            df_frequencia_conjunta.at['Total','Superior']/df_frequencia_conjunta.at['Total','Total']
        ],
        "Total":[
            df_frequencia_conjunta.at['Capital','Total']/df_frequencia_conjunta.at['Capital','Total'],
            df_frequencia_conjunta.at['Interior','Total']/df_frequencia_conjunta.at['Interior','Total'],
            df_frequencia_conjunta.at['Outra','Total']/df_frequencia_conjunta.at['Outra','Total'],
            df_frequencia_conjunta.at['Total','Total']/df_frequencia_conjunta.at['Total','Total']
        ]
    },
    index=['Capital','Interior','Outra','Total']
)
print('Distribuição das proporções do grau de educação segundo cada uma das regiões de procedência:')
df_distribuicao_proporcoes_grau_educacao_segundo_cada_uma_regioes_de_procedência
```

```python
df_frequencia_conjunta_valores_esperados=pd.DataFrame(
    data={
        "Fundamental":[
            [
                df_frequencia_conjunta.at['Capital','Total']*(df_frequencia_conjunta.at['Total','Fundamental']/df_frequencia_conjunta.at['Total','Total']),
                100 * df_frequencia_conjunta.at['Total','Fundamental']/df_frequencia_conjunta.at['Total','Total']
            ],
            [
                df_frequencia_conjunta.at['Interior','Total']*(df_frequencia_conjunta.at['Total','Fundamental']/df_frequencia_conjunta.at['Total','Total']),
                100 * df_frequencia_conjunta.at['Total','Fundamental']/df_frequencia_conjunta.at['Total','Total']
            ],
            [
                df_frequencia_conjunta.at['Outra','Total']*(df_frequencia_conjunta.at['Total','Fundamental']/df_frequencia_conjunta.at['Total','Total']),
                100 * df_frequencia_conjunta.at['Total','Fundamental']/df_frequencia_conjunta.at['Total','Total']
            ],
            [
                df_frequencia_conjunta.at['Total','Fundamental'],
                100 * df_frequencia_conjunta.at['Total','Fundamental']/df_frequencia_conjunta.at['Total','Total']
            ]
        ],
        "Médio":[
            [
                df_frequencia_conjunta.at['Capital','Total']*(df_frequencia_conjunta.at['Total','Médio']/df_frequencia_conjunta.at['Total','Total']),
                100 * df_frequencia_conjunta.at['Total','Médio']/df_frequencia_conjunta.at['Total','Total']
            ],
            [
                df_frequencia_conjunta.at['Interior','Total']*(df_frequencia_conjunta.at['Total','Médio']/df_frequencia_conjunta.at['Total','Total']),
                100 * df_frequencia_conjunta.at['Total','Médio']/df_frequencia_conjunta.at['Total','Total']
            ],
            [
                df_frequencia_conjunta.at['Outra','Total']*(df_frequencia_conjunta.at['Total','Médio']/df_frequencia_conjunta.at['Total','Total']),
                100 * df_frequencia_conjunta.at['Total','Médio']/df_frequencia_conjunta.at['Total','Total']
            ],
            [
                df_frequencia_conjunta.at['Total','Médio'],
                100 * df_frequencia_conjunta.at['Total','Médio']/df_frequencia_conjunta.at['Total','Total']
            ]
        ],
        "Superior":
        [
            [
                df_frequencia_conjunta.at['Capital','Total']*(df_frequencia_conjunta.at['Total','Superior']/df_frequencia_conjunta.at['Total','Total']),
                100 * df_frequencia_conjunta.at['Total','Superior']/df_frequencia_conjunta.at['Total','Total']
            ],
            [
                df_frequencia_conjunta.at['Interior','Total']*(df_frequencia_conjunta.at['Total','Superior']/df_frequencia_conjunta.at['Total','Total']),
                100 * df_frequencia_conjunta.at['Total','Superior']/df_frequencia_conjunta.at['Total','Total']
            ],
            [
                df_frequencia_conjunta.at['Outra','Total']*(df_frequencia_conjunta.at['Total','Superior']/df_frequencia_conjunta.at['Total','Total']),
                100 * df_frequencia_conjunta.at['Total','Superior']/df_frequencia_conjunta.at['Total','Total']
            ],
            [
                df_frequencia_conjunta.at['Total','Superior'],
                100 * df_frequencia_conjunta.at['Total','Superior']/df_frequencia_conjunta.at['Total','Total']
            ]
        ],
        "Total":[
            [
                df_frequencia_conjunta.at['Capital','Total'],
                100 * df_frequencia_conjunta.at['Capital','Total']/df_frequencia_conjunta.at['Capital','Total']
            ],
            [
                df_frequencia_conjunta.at['Interior','Total'],
                100 * df_frequencia_conjunta.at['Interior','Total']/df_frequencia_conjunta.at['Interior','Total']
            ],
            [
                df_frequencia_conjunta.at['Outra','Total'],
                100 * df_frequencia_conjunta.at['Outra','Total']/df_frequencia_conjunta.at['Outra','Total']
            ],
            [
                df_frequencia_conjunta.at['Total','Total'],
                100 * df_frequencia_conjunta.at['Total','Total']/df_frequencia_conjunta.at['Total','Total']
            ]
        ]
    },
    index=['Capital','Interior','Outra','Total']
)
print('Distribuição de frequência conjunta com valores esperados:')
df_frequencia_conjunta_valores_esperados
```

```python
qui_quadrado=0.0
for index in ['Capital','Interior','Outra']:
    for column in ['Fundamental','Médio','Superior']:
        valor_esperado = df_frequencia_conjunta_valores_esperados.at[index, column][0]
        valor_observado = df_frequencia_conjunta.at[index, column]
        
        observado_menos_esperado = valor_observado - valor_esperado
        qui_quadrado += ( observado_menos_esperado**2 )/valor_esperado
        
print('X²:', qui_quadrado)
```

```python
frequencia_total = df_frequencia_conjunta.at['Total','Total']
coeficiente_contingencia_C = math.sqrt( qui_quadrado/(qui_quadrado+frequencia_total) )
print('Coeficiente de contingência C:', coeficiente_contingencia_C)
```
