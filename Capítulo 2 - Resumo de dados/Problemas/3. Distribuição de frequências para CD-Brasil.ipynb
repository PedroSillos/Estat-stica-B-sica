## 3. Distribuição de frequências para CD-Brasil

#### Para o Conjunto de Dados 1 (CD-Brasil), construa a distribuição de frequências para as variáveis população e densidade populacional.

CD-Brasil:


```python
import pandas as pd
```

##### (a) população

```python
de0a10milhoes = 20
de10a20milhoes = 6
de20a30milhoes = 0
de30a40milhoes = 1
total = de0a10milhoes + de10a20milhoes + de20a30milhoes + de30a40milhoes

df = pd.DataFrame(
    {
        "População (em milhões)": ["0 |-- 10", "10 |-- 20", "20 |-- 30", "30 |-- 40", "Total"],
        "Frequência": [de0a10milhoes, de10a20milhoes, de20a30milhoes, de30a40milhoes, total],
        "Proporção": [de0a10milhoes/total, de10a20milhoes/total, de20a30milhoes/total, de30a40milhoes/total, total/total],
        "Porcentagem": [de0a10milhoes*100/total, de10a20milhoes*100/total, de20a30milhoes*100/total, de30a40milhoes*100/total, total*100/total]
    }
)

df
```

##### (b) densidade populacional

```python
de0a10 = 9
de10a50 = 6
de50a150 = 9
de150a500 = 3
total = de0a10 + de10a50 + de50a150 + de150a500

df = pd.DataFrame(
    {
        "Densidade populacional (hab/km quadrados)": ["0 |-- 10", "10 |-- 50", "50 |-- 150", "150 |-- 500", "Total"],
        "Frequência": [de0a10, de10a50, de50a150, de150a500, total],
        "Proporção": [de0a10/total, de10a50/total, de50a150/total, de150a500/total, total/total],
        "Porcentagem": [de0a10*100/total, de10a50*100/total, de50a150*100/total, de150a500*100/total, total*100/total]
    }
)

df
```
