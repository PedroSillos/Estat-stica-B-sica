## 2. Distribuição de frequências

#### Usando os dados da tabela 2.1, construa a distribuição de frequências das variáveis:

Tabela 2.1:


```python
import pandas as pd
```

##### (a) Estado civil

```python
solteiros = 16
casados = 20
total = solteiros + casados

df = pd.DataFrame(
    {
        "Estado Civil": ["solteiro", "casado","Total"],
        "Frequência": [solteiros, casados, total],
        "Proporção": [solteiros/total, casados/total, total/total],
        "Porcentagem": [100*solteiros/total, 100*casados/total, 100*total/total]
    }
)

df
```

##### (b) Região de procedência

```python
interior = 12
capital = 11
outra = 13
total = interior + capital + outra

df = pd.DataFrame(
    {
        "Região de Procedência": ["interior","capital","outra", "Total"],
        "Frequência": [interior, capital, outra, total],
        "Proporção": [interior/total, capital/total, outra/total, total/total],
        "Porcentagem": [100*interior/total, 100*capital/total, 100*outra/total, 100*total/total],
    }
)

df
```

##### (c) Número de filhos dos empregados casados

```python
tem0filhos = 4
tem1filho = 5
tem2filhos = 7
tem3filhos = 3
tem4filhos = 0
tem5filhos = 1
total = tem0filhos + tem1filho + tem2filhos + tem3filhos + tem4filhos + tem5filhos

df = pd.DataFrame(
    {
        "N de filhos": ["0", "1", "2", "3", "4", "5","Total"],
        "Frequência": [tem0filhos, tem1filho, tem2filhos, tem3filhos, tem4filhos, tem5filhos, total],
        "Proporção": [tem0filhos/total, tem1filho/total, tem2filhos/total, tem3filhos/total, tem4filhos/total, tem5filhos/total, total/total],
        "Porcentagem": [100*tem0filhos/total, 100*tem1filho/total, 100*tem2filhos/total, 100*tem3filhos/total, 100*tem4filhos/total, 100*tem5filhos/total, 100*total/total]
    }
)

df
```

##### (d) Idade

```python
de20a25 = 2
de25a30 = 6
de30a35 = 10
de35a40 = 8
de40a45 = 8
de45a50 = 2
total = de20a25 + de25a30 + de30a35 + de35a40 + de40a45 + de45a50

df = pd.DataFrame(
    {
        "Idade": ["20 |-- 25","25 |-- 30","30 |-- 35","35 |-- 40","40 |-- 45","45 |--| 50", "Total"],
        "Frequência": [de20a25, de25a30, de30a35, de35a40, de40a45, de45a50, total],
        "Proporção": [de20a25/total, de25a30/total, de30a35/total, de35a40/total, de40a45/total, de45a50/total, total/total],
        "Porcentagem": [de20a25*100/total, de25a30*100/total, de30a35*100/total, de35a40*100/total, de40a45*100/total, de45a50*100/total, total*100/total]
    }
)

df
```
