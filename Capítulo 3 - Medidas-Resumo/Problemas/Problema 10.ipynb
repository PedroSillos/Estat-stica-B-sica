## Problema 10

#### Para a variável população do CD-Brasil, obtenha q(0,10), q(0,25) ,q(0,50) ,q(0,75), q(0,80) e q(0,95).

```python
import pandas as pd
```

```python
df = pd.read_excel(r'C:\Estatística Básica\Conjunto de Dados.xls', sheet_name='CD-Brasil', header=0, skiprows=[0,1,2,3,12,22,27,31,36,37], decimal='.')
df
```

```python
description = df.describe(percentiles=[.10, .25, .50, .75, .80, .95])
description
```

```python
print('q(0,10) População:','{0:.10g}'.format(description.at['10%','População']))
print('q(0,25) População:','{0:.10g}'.format(description.at['25%','População']))
print('q(0,50) População:','{0:.10g}'.format(description.at['50%','População']))
print('q(0,75) População:','{0:.10g}'.format(description.at['75%','População']))
print('q(0,80) População:','{0:.10g}'.format(description.at['80%','População']))
print('q(0,95) População:','{0:.10g}'.format(description.at['95%','População']))
```
