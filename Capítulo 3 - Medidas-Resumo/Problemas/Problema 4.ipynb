##### (b) Esboce um histograma em que a média e a mediana coincidem. Existe alguma classe de histogramas em que isso sempre acontece?

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
distribuicao_normal=np.random.normal(loc=0, scale=0.95, size=100000)
plt.hist(distribuicao_normal, range=(-4,4), density=True, rwidth=0.9, color='Orange')
plt.show()
```

Para qualquer distribuição simétrica a média e a mediana serão iguais
